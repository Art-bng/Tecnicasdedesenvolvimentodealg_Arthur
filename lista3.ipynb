```python
def calcular_soma_e_media():
    while True:
        try:
            n_str = input("Digite a quantidade de números que você irá informar (N): ")
            n = int(n_str)

            if n <= 0:
                print("A quantidade deve ser um número inteiro positivo.")
                continue
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro para a quantidade.")

    soma = 0

    print(f"\nAgora, digite {n} números:")
    for i in range(1, n + 1):
        while True:
            try:
                numero_str = input(f"Digite o número {i}: ")
                numero = int(numero_str)

                soma += numero
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")

    media = soma / n


    print("\n--- Resultados ---")
    print(f"Quantidade de números lidos (N): {n}")
    print(f"Soma de todos os números: {soma}")
    print(f"Média Aritmética: {media:.2f}")
calcular_soma_e_media()
```

```python

def encontrar_pessoa_mais_velha():

    while True:
        try:
            n_str = input("Digite a quantidade de pessoas que serão registradas (N): ")
            n = int(n_str)

            if n <= 0:
                print("A quantidade deve ser um número inteiro positivo.")
                continue
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro para a quantidade.")
    idade_maxima = -1
    nome_mais_velho = ""

    print(f"\n--- Inserção de {n} Registros ---")

    for i in range(n):
        print(f"\nRegistro {i + 1} de {n}:")

        nome = input("Digite o nome: ")

        while True:
            try:
                idade_str = input("Digite a idade: ")
                idade = int(idade_str)

                if idade < 0:
                    print("A idade não pode ser negativa.")
                    continue
                break
            except ValueError:
                print("Idade inválida. Por favor, digite um número inteiro.")

        if idade > idade_maxima:
            idade_maxima = idade
            nome_mais_velho = nome

    print("\n==================================")
    print("RESULTADO FINAL")
    print(f"A pessoa mais velha é: {nome_mais_velho}")
    print(f"Com a idade de: {idade_maxima} anos")
    print("==================================")
```

```python

def calcular_media_idade_por_genero():
    while True:
        try:
            n_str = input("Digite a quantidade de pessoas que serão registradas (N): ")
            n = int(n_str)

            if n <= 0:
                print("A quantidade deve ser um número inteiro positivo.")
                continue
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    soma_idade_f = 0
    contador_f = 0
    soma_idade_m = 0
    contador_m = 0


    print(f"\n--- Inserção de {n} Registros ---")

    for i in range(n):
        print(f"\nRegistro {i + 1} de {n}:")

        while True:
            sexo = input("Digite o sexo (M para Masculino, F para Feminino): ").upper()
            if sexo in ['M', 'F']:
                break
            print("Sexo inválido. Por favor, digite 'M' ou 'F'.")

        while True:
            try:
                idade_str = input("Digite a idade: ")
                idade = int(idade_str)

                if idade < 0:
                    print("A idade não pode ser negativa.")
                    continue
                break
            except ValueError:
                print("Idade inválida. Por favor, digite um número inteiro.")


        if sexo == 'F':
            soma_idade_f += idade
            contador_f += 1
        elif sexo == 'M':
            soma_idade_m += idade
            contador_m += 1

    if contador_f > 0:
        media_f = soma_idade_f / contador_f
    else:
        media_f = 0

    if contador_m > 0:
        media_m = soma_idade_m / contador_m
    else:
        media_m = 0

    print("\n==================================")
    print("MÉDIAS DE IDADE POR GÊNERO")
    print("==================================")

    print(f"Total de mulheres cadastradas: {contador_f}")
    print(f"Média de idade Feminina: {media_f:.2f} anos")

    print("-" * 34)

    print(f"Total de homens cadastrados: {contador_m}")
    print(f"Média de idade Masculina: {media_m:.2f} anos")

    print("==================================")
```

```python
def encontrar_extremos_dez_numeros():

    maior_numero = None
    menor_numero = None

    quantidade_numeros = 10

    print(f"--- Início da Leitura de {quantidade_numeros} Números ---")

    for i in range(quantidade_numeros):
        while True:
            try:

                entrada_str = input(f"Digite o {i + 1}º número real: ")
                numero_atual = float(entrada_str)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número real (ex: 3.14).")

        if maior_numero is None and menor_numero is None:
            maior_numero = numero_atual
            menor_numero = numero_atual

        elif numero_atual > maior_numero:
            maior_numero = numero_atual

        elif numero_atual < menor_numero:
            menor_numero = numero_atual

    print("\n--- Resultados Finais ---")
    print(f"O MAIOR número digitado foi: {maior_numero}")
    print(f"O MENOR número digitado foi: {menor_numero}")
    print("-------------------------")
```

```python
def ordenar_numeros_ate_zero():

    lista_numeros = []

    print("--- Início da Entrada de Dados ---")
    print("Digite números reais (com casas decimais ou inteiros).")
    print("Para parar e ordenar os números, digite 0 (zero).")

    while True:
        try:
            entrada_str = input("Digite um número: ")

            numero_atual = float(entrada_str)
            if numero_atual == 0:
                print("\nZero (0) detectado. Parando a entrada de dados.")
                break
            lista_numeros.append(numero_atual)

        except ValueError:
            print("Entrada inválida. Por favor, digite apenas números reais.")

    if len(lista_numeros) == 0:
        print("Nenhum número válido foi inserido para ordenação.")
        return
    lista_numeros.sort()

    print("\n--- Números Ordenados (Crescente) ---")

    print(lista_numeros)

    print("--------------------------------------")
```

```python
def calcular_estatisticas_sala():
    idades = []

    print("--- Registro de Idades de Alunos ---")

    while True:
        proximo = input("\nDeseja informar a idade do próximo aluno? (S/N): ").strip().upper()

        if proximo != 'S':
            break
        while True:
            try:
                idade_str = input("Digite a idade do aluno: ")
                idade = int(idade_str)

                if idade < 0 or idade > 120:
                    print("Por favor, digite uma idade válida (entre 0 e 120).")
                    continue

                idades.append(idade)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro para a idade.")


    n = len(idades)
    if n == 0:
        print("\nNenhuma idade foi informada. Encerrando a análise.")
        return

    idade_mais_novo = min(idades)
    idade_mais_velho = max(idades)

    maiores_de_18 = 0
    ate_18_anos = 0

    for idade in idades:
        if idade > 18:
            maiores_de_18 += 1
        else:
            ate_18_anos += 1

    soma_idades = sum(idades)
    media_aritmetica = soma_idades / n
    idades.sort()

    if n % 2 == 1:
        indice_central = (n - 1) // 2
        mediana = idades[indice_central]
    else:
        indice1 = n // 2 - 1
        indice2 = n // 2
        mediana = (idades[indice1] + idades[indice2]) / 2


    print("\n==================================")
    print("        RELATÓRIO DE IDADES       ")
    print("==================================")

    print(f"Total de alunos registrados: {n}")
    print("-" * 34)

    print(f"Idade do aluno mais novo: {idade_mais_novo}")
    print(f"Idade do aluno mais velho: {idade_mais_velho}")
    print("-" * 34)

    print(f"Alunos com MAIS de 18 anos: {maiores_de_18}")
    print(f"Alunos com ATÉ 18 anos: {ate_18_anos}")
    print("-" * 34)

    print(f"Média Aritmética das Idades: {media_aritmetica:.2f}")
    print(f"Mediana das Idades: {mediana:.2f}")
    print("==================================")

if __name__ == "__main__":
    calcular_estatisticas_sala()
```
